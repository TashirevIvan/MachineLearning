Имена признаков:
1. Name - Название видеоигры 
2. Platform - Платформа, на которой игра вышла
3. YearofRelease - Дата выхода
4. Genre - Жанр
5. Publisher - Издатель
6. NASales - Уровень продаж в Северной Америке 
7. EUSales - Уровень продаж в Европе
8. JPSales - Уровень продаж в Японии
9. OtherSales - Уровень продаж в останьных регионах

12. Criticcount - Количество критиков, использованных при составлении рейтинга критиков
13. Userscore - Оценка подписчиков Metacritic
14. Usercount - Количество пользователей, которые дали рейтинг
15. Developer - Разработчик, ответственный за создание игры
16. Rating - ESRB рейтинг

# Обработка пропущенных значений

```python
data_raw.isna()
```

```python
data_raw.isna().sum()
```

```python
data_raw.fillna(data_raw.median(axis = 0), axis=0 , inplace=True)
```

```python
data_raw.isna().sum()
```

```python
data_raw['Name'].fillna(data_raw['Name'].mode().iloc[0], inplace=True)
data_raw['Year_of_Release'].fillna(data_raw['Year_of_Release'].mode().iloc[0], inplace=True)
data_raw['Genre'].fillna(data_raw['Genre'].mode().iloc[0], inplace=True)
data_raw['Publisher'].fillna(data_raw['Publisher'].mode().iloc[0], inplace=True)
data_raw['Developer'].fillna(data_raw['Developer'].mode().iloc[0], inplace=True)
data_raw['Rating'].fillna(data_raw['Rating'].mode().iloc[0], inplace=True)
```

```python
data_raw.isna().sum()
```

```python
data_raw.head(10)
```

# Обработка категориальных значений

```python
data_new = data_raw.loc[:, ['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher',
                            'Global_Sales', 'Critic_Score', 'Critic_Count', 'User_Score',
                           'User_Count', 'Developer', 'Rating', 'binary_Global_Sales']].copy()
```

```python
data_new.head()
```

```python
Runtime_dummies = pd.get_dummies(data_new['Platform'])
```

```python
data_new = pd.concat((data_new, Runtime_dummies), axis=1)
data_new = data_new.drop(['Platform'], axis=1)
```

```python
data_new.head()
```

# Нормализация количественных признаков

```python
data_stand = (data_new - data_new.mean(axis = 0))/data_new.std(axis = 0)
```

```python
data_stand.describe()
```
